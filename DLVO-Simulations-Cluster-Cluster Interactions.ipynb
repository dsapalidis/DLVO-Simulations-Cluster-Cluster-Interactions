```python
## DLVO simulations ##
 
## DLVO simulations for Cluster-Cluster interactions considering long-range potentials (version 4.0) ##

## D.Sapalidis, Zurich, 18.01.2026 ##

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from pathlib import Path

# Paths
input_file = Path(r'C:\Users\dsapa\Desktop\extra\Υπολογισμοί\DLVO Cluster-Cluster Correct\DLVO_ClusterCluster_Input.txt')
output_folder = Path(r'C:\Users\dsapa\Desktop\extra\Υπολογισμοί\DLVO Cluster-Cluster Correct')
output_folder.mkdir(parents=True, exist_ok=True)  

# Input file
try:
    data = pd.read_csv(input_file, delimiter='\t')
    data.columns = data.columns.str.strip()  
except Exception as e:
    print(f"Error reading input file: {e}")
    raise

# Input data
numeric_columns = ['Av0(J)', 'Av(J)', 'scrit(m)', 'ε0(F/m)', 'εr', 'ψagg(V)', 'σψagg(V)', 'κ (1/m)', 'Rg(m)', 'φ']
for col in numeric_columns:
    if col in data.columns:
        data[col] = pd.to_numeric(data[col], errors='coerce')

# Van der Waals potential function
def V_vdW(s, Av0, Av, s_crit, Rg, phi):
    A_131 = Av0 * (1 + 2 * kappa * s) * np.exp(-2 * kappa * s) + Av * (1 + (s/s_crit)**(3/2))**(-2/3)
    return - 4 * (phi**2 / 27) * A_131 * ((3 * Rg**2) / (s + 2*Rg)**2 - (4 * Rg**2) / ((s + 2*Rg)**2 - Rg**2) + (Rg**2) / ((s + 2*Rg)**2 - (4 * Rg**2)))

# Electrostatic potential function
def V_elec(s, epsilon_0, epsilon_r, psi_agg, sigma_psi_agg, kappa, Rg):
    part1 = (psi_agg**2 + sigma_psi_agg**2) * np.log(1 - np.exp(-2 * kappa * s))
    part2 = psi_agg**2 * np.log(np.cosh(kappa * s / 2) / np.sinh(kappa * s / 2))
    return 3.14 * epsilon_0 * epsilon_r * Rg * (part1 + part2)


kBT = 1.38e-23 * 293

# V_max function
def get_Vmax(s_nm, V_norm):
    
    V = np.asarray(V_norm, dtype=float)
    s = np.asarray(s_nm, dtype=float)

    finite = np.isfinite(V)
    if not np.any(finite):
        return np.nan, np.nan

    idx_f = np.argmax(V[finite])
    idx = np.where(finite)[0][idx_f]
    return V[idx], s[idx]

# Calculate potentials and plot
for i in range(len(data)):
    try:
        
        Rg = data.at[i, 'Rg(m)']
        Av0 = data.at[i, 'Av0(J)']
        Av = data.at[i, 'Av(J)']
        s_crit = data.at[i, 'scrit (m)']
        epsilon_0 = data.at[i, 'ε0(F/m)']
        epsilon_r = data.at[i, 'εr']
        psi_agg = data.at[i, 'ψagg(V)']
        sigma_psi_agg = data.at[i, 'σψagg(V)']
        kappa = data.at[i, 'κ (1/m)']
        phi = data.at[i, 'φ']
        
        # Metadata information
        buffer = data.at[i, 'Buffer'] if 'Buffer' in data.columns else 'unknown'
        concentration = data.at[i, 'w/v(mg/ml)'] if 'w/v(mg/ml)' in data.columns else 'unknown'
        sonicated = data.at[i, 'son'] if 'son' in data.columns else 'unknown'
        
        
        filename_base = f"{buffer}_{concentration}_{sonicated}".replace('/', '_').replace('.', '_')

        # Print all values 
        print(f"\nSample {i + 1} Parameters:")
        print(f"Buffer: {buffer}, Concentration: {concentration} mg/ml, Sonicated: {sonicated}")
        print(f"Av0(J): {Av0}, Av(J): {Av}, s_crit: {s_crit}")
        print(f"ε0(F/m): {epsilon_0}, εr: {epsilon_r}, ψagg(V): {psi_agg}, σψagg(V): {sigma_psi_agg}")
        print(f"κ (1/m): {kappa}, Rg(m): {Rg}, φ: {phi}\n")

        # Generate s_values
        s_values = np.linspace(0, 40 * 10**(-9), 500)  # Start at 0 

        # Calculate V_vdW, V_elec, and V_total
        V_vdW_values = V_vdW(s_values, Av0, Av, s_crit, Rg, phi)
        V_elec_values = V_elec(s_values, epsilon_0, epsilon_r, psi_agg, sigma_psi_agg, kappa, Rg)
        V_total = V_vdW_values + V_elec_values

        # Normalize the potentials
        s_norm = s_values * 10**9
        V_norm_vdW = V_vdW_values / kBT
        V_norm_elec = V_elec_values / kBT
        V_norm_total = V_total / kBT
        Vmax, s_at_Vmax = get_Vmax(s_norm, V_norm_total)
        
        

        # Save data
        output_data = pd.DataFrame({
            's [nm]': s_norm,
            'V_vdW/(kBT)': V_norm_vdW,
            'V_elec/(kBT)': V_norm_elec,
            'V_total/(kBT)': V_norm_total
        })
        output_data_path = output_folder / f'{filename_base}_Cluster_Potential_data.csv'
        output_data.to_csv(output_data_path, index=False, float_format='%.8f')
        print(f"Data saved to {output_data_path}")
        print(f"Vmax (V_total/kBT) = {Vmax:.6g} at s = {s_at_Vmax:.6g} nm")

        # Plotting
        plt.figure(figsize=(10, 6))
        plt.plot(s_norm, V_norm_vdW, 'r--', label=r'$V_{vdW}/(k_B T)$')
        plt.plot(s_norm, V_norm_elec, 'g--', label=r'$V_{elec}/(k_B T)$')
        plt.plot(s_norm, V_norm_total, 'k-', label=r'$V_{tot}/(k_B T)$')
        plt.xlabel(r'$s/nm$')
        plt.ylabel(r'$V/(k_B T)$')
        plt.title(f"DLVO Potential for {buffer}, {concentration} mg/ml, Sonicated: {sonicated}")
        plt.legend()
        plt.grid(True)
        plt.show()
        
        # Plotting
        #plt.figure(figsize=(10, 6))
        #plt.plot(s_norm, V_norm_vdW, 'r--', label=r'$V_{vdW}/(k_B T)$')
        #plt.xlabel(r'$s/nm$')
        #plt.ylabel(r'$V/(k_B T)$')
        #plt.title(f"DLVO Potential for {buffer}, {concentration} mg/ml, Sonicated: {sonicated}")
        #plt.legend()
        #plt.grid(True)
        #plt.show()
        


    
    except KeyError as e:
        print(f"Missing value in required columns for sample {i + 1}: {e}")
        continue
```
