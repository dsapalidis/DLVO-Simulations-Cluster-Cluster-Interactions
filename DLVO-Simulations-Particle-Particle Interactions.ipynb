```python
## DLVO simulations ##
 
## DLVO simulations for Particle-Particle interactions considering long-range potentials (version 4.0) ##

## D.Sapalidis, Empa, Center for X-ray Analytics, St. Gallen, 18.01.2026 ##

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from pathlib import Path

input_folder = Path(r'C:\Users\dsapa\Desktop\extra\Υπολογισμοί\DLVO Particle-Particle Correct')
output_folder = Path(r'C:\Users\dsapa\Desktop\extra\Υπολογισμοί\DLVO Particle-Particle Correct')
output_folder.mkdir(parents=True, exist_ok=True)

input_file = input_folder / 'DLVO_ParticleParticle_Input.txt'
data = pd.read_csv(input_file, delimiter='\t', dtype=str)

data.columns = data.columns.str.strip()

numeric_columns = [
    'Av0(J)', 'Av(J)', 's_crit (m)', 'Rpr(m)', 'ε0(F/m)', 'εr', 'ψ0(V)', 'κ (1/m)', 'Y',
    'kB(J/K)', 'T(K)', 'e[C]', 'v solv(m3/mol)', 'Φp', 'χ', 'w(m)',
    'MWp(kg/mol)', 'ρp (kg/m3)'
]

for col in numeric_columns:
    if col in data.columns:
        data[col] = pd.to_numeric(data[col], errors='coerce')
    else:
        print(f"Column '{col}' not found in the DataFrame.")

print("Preview of the DataFrame after conversion:")
print(data.head())

if 'Rpr(m)' in data.columns:
    Rmin = data['Rpr(m)'].dropna().astype(float).min()
    Rmax = data['Rpr(m)'].dropna().astype(float).max()
    r_values = np.linspace(2.0 * Rmin * (1.0 + 1e-8), 10.0 * Rmax, 500)
else:
    print("Cannot calculate r_values because 'Rpr(m)' column is missing.")
    r_values = None

def A131_retarded(r, Av0, Av, s_crit, R_pr, kappa, eps_safe=1e-30):
    s = r - 2.0 * R_pr
    s = np.maximum(s, 0.0)
    s_crit_eff = max(float(s_crit), eps_safe)
    F = (1.0 + (s / s_crit_eff) ** 1.5) ** (-2.0 / 3.0)
    return Av0 * (1.0 + 2.0 * kappa * s) * np.exp(-2.0 * kappa * s) + Av * F

def get_Vmax(r_norm, V_norm):
    V = np.asarray(V_norm, dtype=float)
    r = np.asarray(r_norm, dtype=float)
    finite = np.isfinite(V)
    if not np.any(finite):
        return np.nan, np.nan
    idx_f = np.argmax(V[finite])
    idx = np.where(finite)[0][idx_f]
    return V[idx], r[idx]

if r_values is not None:
    for index, row in data.iterrows():
        try:
            Av0 = float(row['Av0(J)'])
            Av = float(row['Av(J)'])
            s_crit = float(row['s_crit (m)'])
            R_pr = float(row['Rpr(m)'])
            epsilon_0 = float(row['ε0(F/m)'])
            epsilon_r = float(row['εr'])
            psi_0 = float(row['ψ0(V)'])
            kappa = float(row['κ (1/m)'])
            Y = float(row['Y'])
            k_B = float(row['kB(J/K)'])
            T = float(row['T(K)'])
            e = float(row['e[C]'])
            v_solv = float(row['v solv(m3/mol)'])
            Phi_p = float(row['Φp'])
            chi = float(row['χ'])
            w = float(row['w(m)'])
            MW_p = float(row['MWp(kg/mol)'])
            rho_p = float(row['ρp (kg/m3)'])
        except ValueError as e:
            print(f"Skipping sample {index + 1} due to ValueError: {e}")
            continue

        print(f"Av0 = {Av0}, Av = {Av}, s_crit = {s_crit}, R_pr = {R_pr}, epsilon_0 = {epsilon_0}, epsilon_r = {epsilon_r}, psi_0 = {psi_0}, kappa = {kappa}")
        print(f"Y = {Y}, k_B = {k_B}, T = {T}, e = {e}, v_solv = {v_solv}, Phi_p = {Phi_p}, chi = {chi}, w = {w}, MW_p = {MW_p}, rho_p = {rho_p}")

        if (R_pr <= 0 or epsilon_0 <= 0 or k_B <= 0 or e <= 0 or v_solv <= 0 or w <= 0 or rho_p <= 0 or
            kappa <= 0 or s_crit <= 0):
            print(f"Skipping sample {index + 1} due to invalid critical parameters.")
            continue

        def V_ndW_retarded(r, Av0, Av, s_crit, R_pr, kappa):
            with np.errstate(divide='ignore', invalid='ignore', over='ignore'):
                A_r = A131_retarded(r, Av0, Av, s_crit, R_pr, kappa)
                denom = (r**2 - 4.0 * R_pr**2)
                return -(A_r / 6.0) * (
                    (2.0 * R_pr**2) / denom +
                    (2.0 * R_pr**2) / (r**2) +
                    np.log(denom / (r**2))
                )

        def V_elec(r, epsilon_0, epsilon_r, R_pr, psi_0, kappa, Y, k_B, T, e):
            with np.errstate(divide='ignore', invalid='ignore', over='ignore'):
                if kappa * R_pr >= 5:
                    return 2 * np.pi * epsilon_0 * epsilon_r * R_pr * psi_0**2 * np.log(1 + np.exp(-kappa * (r - 2 * R_pr)))
                else:
                    return 4 * np.pi * epsilon_0 * epsilon_r * R_pr**2 * Y**2 * (k_B * T / e)**2 * np.exp(-kappa * (r - 2 * R_pr)) / r

        def V_osm(r, R_pr, w, k_B, T, v_solv, Phi_p, chi):
            with np.errstate(divide='ignore', invalid='ignore', over='ignore'):
                result = np.zeros_like(r)
                s = r - 2 * R_pr
                mask1 = (s >= 2 * w)
                mask2 = (w < s) & (s < 2 * w)
                mask3 = (s < w)
                result[mask1] = 0.0
                result[mask2] = (4 * np.pi * k_B * T * R_pr / v_solv) * Phi_p**2 * (0.5 - chi) * (w - s[mask2] / 2)**2
                result[mask3] = (4 * np.pi * k_B * T * R_pr / v_solv) * Phi_p**2 * w**2 * ((s[mask3] / (2 * w)) - 0.25 - np.log(s[mask3] / w))**2
                return result

        def V_elas(r, R_pr, w, rho_p, MW_p, Phi_p, k_B, T):
            with np.errstate(divide='ignore', invalid='ignore', over='ignore'):
                x = (r - 2 * R_pr) / w
                return (2 * np.pi * k_B * T * R_pr * w**2 * rho_p / MW_p) * Phi_p * (
                    x * np.log(x * ((3 - x) / 2)**2) - 6 * np.log((3 - x) / 2) + 3 * (1 - x)
                )

        V_ndW_values = V_ndW_retarded(r_values, Av0, Av, s_crit, R_pr, kappa)
        V_elec_values = V_elec(r_values, epsilon_0, epsilon_r, R_pr, psi_0, kappa, Y, k_B, T, e)
        V_osm_values = V_osm(r_values, R_pr, w, k_B, T, v_solv, Phi_p, chi)
        V_elas_values = V_elas(r_values, R_pr, w, rho_p, MW_p, Phi_p, k_B, T)

        V_total = V_ndW_values + V_elec_values

        r_norm = r_values / R_pr
        V_norm_ndW = V_ndW_values / (k_B * T)
        V_norm_elec = V_elec_values / (k_B * T)
        V_norm_osm = V_osm_values / (k_B * T)
        V_norm_elas = V_elas_values / (k_B * T)
        V_norm_total = V_total / (k_B * T)

        Vmax, r_at_Vmax = get_Vmax(r_norm, V_norm_total)

        buffer = row.get('Buffer', 'unknown')
        sonication = row.get('son', 'unknown')
        concentration = row.get('w/v(mg/ml)', 'unknown')

        plt.figure(figsize=(10, 6))
        plt.plot(r_norm, V_norm_ndW, 'r--', label=r'$V_{ndW}/(k_B T)$', alpha=0.5)
        plt.plot(r_norm, V_norm_elec, 'g--', label=r'$V_{elec}/(k_B T)$', alpha=0.5)
        plt.plot(r_norm, V_norm_total, 'k-', label=r'$V_{tot}/(k_B T)$', alpha=0.5)

        plt.xlim([2, 4])
        plt.ylim([-10, 15])

        plt.xlabel(r'$r/R_{pr}$')
        plt.ylabel(r'$V/(k_B T)$')
        plt.title(f'Normalized Potential vs Normalized Distance {buffer}_{sonication}_{concentration}'.replace('/', '_').replace('.', '_'))
        plt.legend()
        plt.grid(True)

        filename_base = f"{buffer}_{sonication}_{concentration}".replace('/', '_').replace('.', '_')
        output_image_path = output_folder / f'{filename_base}_V_T_vs_r_Rpr.png'
        plt.savefig(output_image_path)
        plt.show()

        output_data = pd.DataFrame({
            'r/Rpr': r_norm,
            'V_ndW/(kBT)': V_norm_ndW,
            'V_elec/(kBT)': V_norm_elec,
            'V_osm/(kBT)': V_norm_osm,
            'V_elas/(kBT)': V_norm_elas,
            'V_total/(kBT)': V_norm_total
        })
        output_data_path = output_folder / f'{filename_base}_data.csv'
        output_data.to_csv(output_data_path, index=False)

        print(f"Vmax (V_total/kBT) = {Vmax:.6g} at r/Rpr = {r_at_Vmax:.6g}")

print("The plots and data files have been created and saved.")
```
